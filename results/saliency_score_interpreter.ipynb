```python
import json
import numpy as np
```

```python
# Load the JSON file line by line
data = []
with open("results_with_saliency_final.json", "r", encoding="utf-8") as file:
    for line in file:
        data.append(json.loads(line))
```

```python
# Iterate through each data point
for entry in data:
    context = entry["context"]
    choices = entry["choices"]
    predicted_choice = entry["predicted_choice"]
    predicted_choice_score = None
    context_saliency_scores = {}
    choice_saliency_scores = {}

    context_sal_scores = []
    context_sal_tokens = []
    for saliency_entry in entry.get("saliency_data", []):
        context_sal_score = []
        if saliency_entry["choice_idx"] == predicted_choice:
            predicted_choice_score = saliency_entry["score"]
            for token in saliency_entry["context_saliency"]:
                context_sal_tokens.append(token[0])
        # Compute average of context saliency scores for each token
        for token in saliency_entry["context_saliency"]:
            context_sal_score.append(token[1])
        context_sal_scores.append(context_sal_score)
        # Extract saliency scores for the predicted choice
        for token in saliency_entry["choice_saliency"]:
            choice_saliency_scores[token[0]] = token[1]
    # Average context scores
    context_sal_scores = np.array(context_sal_scores)
    context_sal_scores = np.mean(context_sal_scores, axis=0)
    # Combine scores & tokens
    for i in range(len(context_sal_tokens)):
        context_saliency_scores[context_sal_tokens[i]] = context_sal_scores[i]

    # Print the results
    print(f"Context: {context}")
    print(f"Choices: {choices}")
    print(f"Predicted Choice Index: {predicted_choice}")
    print(f"Predicted Choice Score: {predicted_choice_score}")
    print("Context Saliency Scores (Token: Average Score):")
    for token, score in context_saliency_scores.items():
        token = token.replace("\u0120", "")
        print(f"  {token}: {score}")
    print("Choice Saliency Scores (Token: Score):")
    for token, score in choice_saliency_scores.items():
        token = token.replace("\u0120", "")
        print(f"  {token}: {score}")
    print("\n" + "-"*80 + "\n")
    
    # Open a file to write the output
    with open("output.txt", "a", encoding="utf-8") as file:
        # Redirect the print statements to the file
        print(f"Context: {context}", file=file)
        print(f"Choices: {choices}", file=file)
        print(f"Predicted Choice Index: {predicted_choice}", file=file)
        print(f"Predicted Choice Score: {predicted_choice_score}", file=file)
        print("Context Saliency Scores (Token: Average Score):", file=file)
        for token, score in context_saliency_scores.items():
            token = token.replace("\u0120", "")
            print(f"  {token}: {score}", file=file)
        print("Choice Saliency Scores (Token: Score):", file=file)
        for token, score in choice_saliency_scores.items():
            token = token.replace("\u0120", "")
            print(f"  {token}: {score}", file=file)
        print("\n" + "-"*80 + "\n", file=file)
```
