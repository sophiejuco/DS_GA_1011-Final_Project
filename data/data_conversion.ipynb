```python
import pandas as pd
import json
import random
```

```python
# Load the data from Excel file
excel_file = 'SAT3.xlsx'  # Replace with your file path
#csv_file = 'SAT2.csv'
data = pd.read_excel(excel_file)
#data = pd.read_csv(csv_file)
```

```python
# Define the structure and convert each row to the specified JSON format
formatted_data = [
    {
        "context": row["prompt"],
        "question": row["question"],
        "answerA": row["mc_a"],
        "answerB": row["mc_b"],
        "answerC": row["mc_c"],
        "answerD": row["mc_d"],
        "correct": str(row["answer"]).upper()  # Convert the answer to uppercase
    }
    for _, row in data.iterrows()
]
```

```python
formatted_data
```

```python
# Shuffle the data to ensure random distribution in splits
random.shuffle(formatted_data)
```

```python
# Calculate the split sizes
total = len(formatted_data)
train_size = int(0.7 * total)  # 70% for training
dev_size = int(0.1 * total)     # 10% for development
test_size = total - train_size - dev_size  # 20% for testing
```

```python
# Split the data
train_data = formatted_data[:train_size]
dev_data = formatted_data[train_size:train_size + dev_size]
test_data = formatted_data[train_size + dev_size:]
```

```python
# Function to save data in the specified line-by-line format
def save_json_line_by_line(data, filename):
    with open(filename, "w") as file:
        for entry in data:
            file.write(json.dumps(entry) + "\n")
```

```python
# Save each split to a JSON file with one object per line
save_json_line_by_line(train_data, "standardizedtest_v3_trn.jsonl")
save_json_line_by_line(dev_data, "standardizedtest_v3_dev.jsonl")
save_json_line_by_line(test_data, "standardizedtest_v3_tst.jsonl")
```
